# SAAM Part I — Minimum-Variance Portfolio (EUR)

Rolling 10yr estimation window, annual rebalancing, long-only min-var vs value-weighted benchmark.  
CO2 scope 1+2 filter applied. Out-of-sample period: Jan 2014 – Dec 2025.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from scipy.optimize import minimize
import warnings
warnings.filterwarnings('ignore')
```

## Setup & data loading

```python
REGION = 'EUR'
START_YR = 2013
END_YR   = 2024
WIN_LEN  = 10        # estimation window in years
MIN_OBS  = 36
STALE_THR = 0.30
LOW_FLOOR = 0.50
LW_FLOOR  = 0.01     # min shrinkage intensity

DATA = 'Data_2026/'
TPL  = 'Data_2026/Template_for_Part_I-SAAM.xlsx'
OUT  = 'Output_2026/'
```

```python
static = pd.read_excel(DATA + 'Static_2025.xlsx')
ri_m_raw = pd.read_excel(DATA + 'DS_RI_T_USD_M_2025.xlsx', sheet_name='RI')
mv_m_raw = pd.read_excel(DATA + 'DS_MV_T_USD_M_2025.xlsx', sheet_name='MV')
ri_y_raw = pd.read_excel(DATA + 'DS_RI_T_USD_Y_2025.xlsx', sheet_name='RI')
mv_y_raw = pd.read_excel(DATA + 'DS_MV_T_USD_Y_2025.xlsx', sheet_name='MV')
co2_s1_raw = pd.read_excel(DATA + 'DS_CO2_SCOPE_1_Y_2025.xlsx', sheet_name='Scope1')
co2_s2_raw = pd.read_excel(DATA + 'DS_CO2_SCOPE_2_Y_2025.xlsx', sheet_name='Scope2')
rev_raw = pd.read_excel(DATA + 'DS_REV_Y_2025.xlsx', sheet_name='REV')
rf_raw  = pd.read_excel(DATA + 'Risk_Free_Rate_2025.xlsx',
                        sheet_name='F-F_Research_Data_Factors')

print(f'Static shape: {static.shape}')
print(f'RI monthly:   {ri_m_raw.shape}')
```

## Datastream cleanup

The raw Datastream exports have some quirks — error rows marked with `ER:`, mixed column types, etc.  
This cell strips all that out and keeps only valid EUR ISINs.

```python
def clean_ds(df, keep_isins):
    df = df.dropna(subset=['ISIN']).copy()
    df = df[~df['ISIN'].astype(str).str.contains('ER:', na=False)]
    df = df[df['ISIN'].isin(keep_isins)]
    names = df.set_index('ISIN')['NAME']
    cols = [c for c in df.columns if c not in ('NAME', 'ISIN') and c is not None]
    data = df.set_index('ISIN')[cols].apply(pd.to_numeric, errors='coerce')
    # normalise col types — timestamps for monthly, ints for yearly
    new_cols = []
    for c in data.columns:
        if hasattr(c, 'year') and not isinstance(c, (int, np.integer)):
            new_cols.append(pd.Timestamp(c))
        else:
            new_cols.append(c)
    data.columns = new_cols
    return data, names
```

```python
eur_isins = set(static.loc[static['Region'] == REGION, 'ISIN'])
print(f'{len(eur_isins)} EUR ISINs in static file')

ri_m, firm_names = clean_ds(ri_m_raw, eur_isins)
mv_m, _          = clean_ds(mv_m_raw, eur_isins)
ri_y, _          = clean_ds(ri_y_raw, eur_isins)
mv_y, _          = clean_ds(mv_y_raw, eur_isins)
co2_s1, _        = clean_ds(co2_s1_raw, eur_isins)
co2_s2, _        = clean_ds(co2_s2_raw, eur_isins)
rev, _           = clean_ds(rev_raw, eur_isins)

isin_name = firm_names.to_dict()
print(f'{ri_m.shape[0]} EUR firms after cleaning')
```

## Date ranges & helper functions

```python
monthly_all = sorted([c for c in ri_m.columns
                       if isinstance(c, pd.Timestamp)
                       and pd.Timestamp('2000-01-01') <= c <= pd.Timestamp('2025-12-31')])

annual_all = sorted([c for c in ri_y.columns if isinstance(c, (int, np.integer))])

# trim everything to these date ranges
ri_m = ri_m[monthly_all]
mv_m = mv_m[[c for c in monthly_all if c in mv_m.columns]]
ri_y = ri_y[[c for c in annual_all if c in ri_y.columns]]
mv_y = mv_y[[c for c in annual_all if c in mv_y.columns]]
co2_s1 = co2_s1[[c for c in annual_all if c in co2_s1.columns]]
co2_s2 = co2_s2[[c for c in annual_all if c in co2_s2.columns]]
rev = rev[[c for c in annual_all if c in rev.columns]]
```

```python
def months_in_year(yr):
    return [c for c in monthly_all if c.year == yr]

def estim_window(yr):
    """10-year rolling window ending Dec of yr."""
    return [c for c in monthly_all
            if (yr - WIN_LEN + 1, 1) <= (c.year, c.month) <= (yr, 12)]

# sanity check
ew = estim_window(2013)
print(f'Window for 2013: {len(ew)} months, {ew[0]:%Y-%m} to {ew[-1]:%Y-%m}')
assert len(ew) == 120
```

## Risk-free rate

Fama-French file gives an annual rate in percent. I compound it down to a monthly decimal.

```python
rf_raw.columns = ['YYYYMM', 'RF_pct']
rf_raw = rf_raw.dropna(subset=['YYYYMM'])
rf_raw['date'] = pd.to_datetime(rf_raw['YYYYMM'].astype(int).astype(str), format='%Y%m')
rf_raw['date'] = rf_raw['date'] + pd.offsets.MonthEnd(0)

# monthly rate via compounding
rf_mon = ((1 + rf_raw.set_index('date')['RF_pct'] / 100) ** (1/12) - 1).squeeze()
rf_mon.name = 'RF'

print(f'RF covers {rf_mon.index.min():%Y-%m} to {rf_mon.index.max():%Y-%m}')
print(f'Sample value 2014-01: {rf_mon.loc["2014-01"].values[0]:.6f}')
```

## Price cleaning

A few things to handle here:
- tiny RI values (< 0.50) are suspect → NaN them out
- forward-fill gaps *within* a stock's active life (don't fill before first or after last valid)
- detect delistings and assign −100% return at the delisting date

```python
# kill low prices
n_low = ((ri_m > 0) & (ri_m < LOW_FLOOR)).sum().sum()
ri_m[ri_m < LOW_FLOOR] = np.nan
print(f'Zapped {n_low} sub-{LOW_FLOOR} price observations')
```

```python
def ffill_internal(row):
    """Only fill gaps between first and last valid obs."""
    fv, lv = row.first_valid_index(), row.last_valid_index()
    if fv is None or lv is None:
        return row
    idx = list(row.index)
    s, e = idx.index(fv), idx.index(lv)
    row.iloc[s:e+1] = row.iloc[s:e+1].ffill()
    return row

before_nans = ri_m.isna().sum().sum()
ri_m = ri_m.apply(ffill_internal, axis=1)
filled = before_nans - ri_m.isna().sum().sum()
print(f'Forward-filled {filled} internal gaps')
```

```python
# find stocks that disappear before end of sample → probable delistings
cutoff = pd.Timestamp('2025-12-31')
delist_dates = {}
for isin in ri_m.index:
    lv = ri_m.loc[isin].last_valid_index()
    if lv is not None and isinstance(lv, pd.Timestamp) and lv < cutoff:
        delist_dates[isin] = lv

print(f'{len(delist_dates)} firms with early last price (potential delistings)')
```

```python
# compute monthly returns from RI levels, then inject -100% at delisting
ret_m = ri_m.pct_change(axis=1)

for isin, dd in delist_dates.items():
    if dd not in monthly_all:
        continue
    pos = monthly_all.index(dd)
    if pos + 1 < len(monthly_all):
        ret_m.at[isin, monthly_all[pos + 1]] = -1.0
        # everything after the -100% is meaningless
        for k in range(pos + 2, len(monthly_all)):
            if monthly_all[k] in ret_m.columns:
                ret_m.at[isin, monthly_all[k]] = np.nan

ret_m = ret_m.iloc[:, 1:]   # first col has no return
print(f'Delisting returns applied: {(ret_m == -1.0).sum().sum()}')
```

```python
# forward-fill annual CO2 & revenue (use last available year if current is missing)
co2_s1 = co2_s1.ffill(axis=1)
co2_s2 = co2_s2.ffill(axis=1)
rev = rev.ffill(axis=1)
```

## Universe construction

For each year Y (2013–2024) we select the investable set based on:  
1. Valid RI price in Dec of year Y  
2. At least 36 non-missing monthly returns in the 10yr window  
3. Not stale (zero-return fraction < 30%)  
4. Both CO2 scope 1 and scope 2 available for year Y

```python
def build_universe(yr):
    dec_cols = [c for c in monthly_all if c.year == yr and c.month == 12]
    if not dec_cols:
        return []
    dec = dec_cols[0]

    win = estim_window(yr)
    win_ret_cols = [c for c in win if c in ret_m.columns]
    R_win = ret_m.reindex(columns=win_ret_cols)

    eligible = []
    for isin in ri_m.index:
        # need a valid Dec price
        if pd.isna(ri_m.at[isin, dec]):
            continue
        if isin not in R_win.index:
            continue

        row = R_win.loc[isin]
        n_ok = row.notna().sum()
        if n_ok < MIN_OBS:
            continue

        # stale check
        n_zero = ((row == 0) | (row.abs() < 1e-10)).sum()
        if n_zero / n_ok > STALE_THR:
            continue

        # CO2 filter — need both scopes
        ok_s1 = (yr in co2_s1.columns and isin in co2_s1.index
                 and pd.notna(co2_s1.at[isin, yr]))
        ok_s2 = (yr in co2_s2.columns and isin in co2_s2.index
                 and pd.notna(co2_s2.at[isin, yr]))
        if not (ok_s1 and ok_s2):
            continue

        eligible.append(isin)
    return eligible
```

```python
universe = {}
for yr in range(START_YR, END_YR + 1):
    universe[yr] = build_universe(yr)
    print(f'{yr}: {len(universe[yr]):3d} firms')
```

## Covariance estimation

Two-step approach:
1. **Pairwise-complete sample covariance** — each pair (i,j) uses only their jointly observed months, which avoids throwing away data when stocks have different listing histories.  
   Var is ML (divide by n), and we reconstruct Cov from the pairwise correlations and own-variances.
2. **Shrinkage toward a constant-correlation target** — the OAS formula (Chen et al. 2010) picks the shrinkage intensity, it also handles the missing-data situation cleanly.  
   Then clip eigenvalues as a safety net to guarantee PSD.

```python
def estimate_cov(isins, win_cols):
    win_in = [c for c in win_cols if c in ret_m.columns]
    R = ret_m.loc[isins, win_in]
    N, T = len(isins), len(win_in)

    mu = R.mean(axis=1).values

    R_vals = R.values                    # (N, T)
    ok = ~np.isnan(R_vals)               # mask
    mu_full = np.nanmean(R_vals, axis=1) # per-firm mean

    # ML variance per firm
    dm = R_vals - mu_full[:, None]        # demeaned, NaN propagates
    dm0 = np.where(ok, dm, 0.0)          # zero out NaN slots
    n_obs = ok.sum(axis=1)
    var = np.where(n_obs > 1, np.sum(dm0**2, axis=1) / n_obs, 0.0)

    # --- pairwise correlations (vectorised) ---
    R0 = np.where(ok, R_vals, 0.0)
    ok_f = ok.astype(np.float64)

    cnt = ok_f @ ok_f.T                  # common obs count
    safe_cnt = np.maximum(cnt, 1)

    sum_i = R0 @ ok_f.T
    sum_j = ok_f @ R0.T
    mi = sum_i / safe_cnt
    mj = sum_j / safe_cnt

    cross = R0 @ R0.T
    cov_pw = cross / safe_cnt - mi * mj

    # pairwise std devs for correlation
    ssq_i = (R0**2) @ ok_f.T
    vi = np.maximum(ssq_i / safe_cnt - mi**2, 0.0)
    ssq_j = ok_f @ (R0**2).T
    vj = np.maximum(ssq_j / safe_cnt - mj**2, 0.0)

    denom = np.sqrt(vi * vj)
    corr_pw = np.where(denom > 1e-20, cov_pw / denom, 0.0)

    # rebuild cov = rho_ij * sig_i * sig_j
    sig = np.sqrt(var)
    S = corr_pw * np.outer(sig, sig)
    np.fill_diagonal(S, var)
    S = (S + S.T) / 2   # symmetry

    # --- Ledoit-Wolf shrinkage to constant-correlation target ---
    sd = np.sqrt(np.diag(S))
    sd_safe = np.where(sd > 1e-20, sd, 1e-20)
    C = S / np.outer(sd_safe, sd_safe)
    np.fill_diagonal(C, 1.0)
    C = np.clip(C, -1, 1)

    rho_avg = (C.sum() - N) / (N * (N - 1))

    # target F: constant correlation, own variances on diagonal
    F = rho_avg * np.outer(sd, sd)
    np.fill_diagonal(F, var)

    # OAS shrinkage intensity
    trS  = np.trace(S)
    trS2 = np.trace(S @ S)
    T_eff = max(np.median(cnt[np.triu_indices(N, k=1)]), 2)

    num = (1 - 2/N) * trS2 + trS**2
    den = (T_eff + 1 - 2/N) * (trS2 - trS**2 / N)

    if abs(den) < 1e-20:
        delta = 0.5
    else:
        delta = max(min(num / den, 1.0), LW_FLOOR)

    Sigma = delta * F + (1 - delta) * S

    # eigenvalue clipping for PSD guarantee
    evals, evecs = np.linalg.eigh(Sigma)
    if evals[0] < 1e-10:
        evals = np.maximum(evals, 1e-10)
        Sigma = evecs @ np.diag(evals) @ evecs.T
        Sigma = (Sigma + Sigma.T) / 2

    return mu, Sigma
```

## Min-variance optimisation

Long-only (α ≥ 0), fully invested (Σα = 1).  
Warm-started from the previous year's drifted weights — i.e. what the portfolio actually looks like after 12 months of price moves, before we rebalance.

```python
drifted_w = {}   # stores end-of-year drifted weights for warm start

def solve_minvar(Sigma, isins, yr):
    N = Sigma.shape[0]
    # warm start from drifted portfolio if available
    if yr in drifted_w:
        prev = drifted_w[yr]
        w0 = np.array([prev.get(i, 0.0) for i in isins])
        s = w0.sum()
        w0 = w0 / s if s > 0 else np.ones(N) / N
    else:
        w0 = np.ones(N) / N

    res = minimize(
        lambda w: float(w @ Sigma @ w),
        w0,
        jac=lambda w: 2.0 * (Sigma @ w),
        method='SLSQP',
        bounds=[(0, None)] * N,
        constraints={'type': 'eq', 'fun': lambda w: w.sum() - 1.0},
        options={'ftol': 1e-10, 'maxiter': 1000},
    )
    if not res.success:
        print(f'  !! optimizer warning yr={yr}: {res.message}')
    return res.x
```

### Out-of-sample return handling

During the OOS year a stock might delist (price vanishes). If last month it had a valid price and now it doesn't, that's a delisting → −100%. If it never had a price (leading NaN), treat as 0%.

```python
def get_oos_returns(eligible, oos_months):
    R = ret_m.loc[eligible].reindex(columns=oos_months).copy()
    for isin in eligible:
        gone = False
        for t in oos_months:
            if gone:
                R.at[isin, t] = np.nan
                continue
            if pd.isna(R.at[isin, t]):
                t_pos = monthly_all.index(t)
                prev = monthly_all[t_pos - 1] if t_pos > 0 else None
                had_price = (prev is not None and prev in ri_m.columns
                             and isin in ri_m.index
                             and pd.notna(ri_m.at[isin, prev]))
                if had_price:
                    R.at[isin, t] = -1.0
                    gone = True
                else:
                    R.at[isin, t] = 0.0
    R = R.fillna(0.0)
    return R
```

## Main loop: estimate → optimise → track OOS returns

```python
mv_weights = {}   # yr → Series of optimal weights
mv_ret = {}       # yr+1 → Series of monthly portfolio returns

for yr in range(START_YR, END_YR + 1):
    firms = universe[yr]
    N = len(firms)
    if N == 0:
        print(f'{yr}: empty universe, skipping')
        continue

    mu, Sig = estimate_cov(firms, estim_window(yr))
    w = solve_minvar(Sig, firms, yr)
    mv_weights[yr] = pd.Series(w, index=firms)

    ea_vol = np.sqrt(float(w @ Sig @ w) * 12) * 100

    # track OOS returns with weight drift
    nxt = months_in_year(yr + 1)
    R_nxt = get_oos_returns(firms, nxt)
    ww = w.copy()
    monthly_pr = []
    for t in nxt:
        rt = R_nxt[t].values
        rp = float(ww @ rt)
        monthly_pr.append(rp)
        ww = ww * (1 + rt) / max(1 + rp, 1e-12)

    mv_ret[yr + 1] = pd.Series(monthly_pr, index=nxt)
    drifted_w[yr + 1] = dict(zip(firms, ww))

    n_nz = (w > 1e-6).sum()
    print(f'{yr}: {N:3d} firms | {n_nz:3d} non-zero weights | ex-ante vol {ea_vol:.2f}%')

rp_mv = pd.concat(mv_ret).droplevel(0).sort_index()
rp_mv.index = pd.DatetimeIndex(rp_mv.index)
```

## Value-weighted benchmark

```python
vw_ret = {}
for yr in range(START_YR, END_YR + 1):
    firms = universe[yr]
    nxt = months_in_year(yr + 1)
    R_oos = get_oos_returns(firms, nxt)
    pr = []
    for t in nxt:
        pos = monthly_all.index(t) if t in monthly_all else None
        if pos is None or pos == 0:
            pr.append(np.nan)
            continue
        prev_t = monthly_all[pos - 1]
        cap = (mv_m.loc[firms, prev_t].fillna(0)
               if prev_t in mv_m.columns
               else pd.Series(0.0, index=firms))
        tot = cap.sum()
        if tot <= 0:
            pr.append(0.0)
            continue
        rt = R_oos[t].values if t in R_oos.columns else np.zeros(len(firms))
        pr.append(float((cap / tot).values @ rt))
    vw_ret[yr + 1] = pd.Series(pr, index=nxt)

rp_vw = pd.concat(vw_ret).droplevel(0).sort_index()
rp_vw.index = pd.DatetimeIndex(rp_vw.index)
```

## Performance comparison

```python
def perf_stats(rp, rf_s, name):
    rp = rp.dropna()
    rf = rf_s.reindex(rp.index).ffill().fillna(0)
    T = len(rp)
    mu_a  = 12 * rp.mean()                         # arith annualised
    mu_g  = (1 + rp).prod() ** (12 / T) - 1        # geom annualised
    vol   = rp.std() * np.sqrt(12)
    rf_a  = 12 * rf.mean()
    sr    = (mu_a - rf_a) / vol
    cum   = (1 + rp).cumprod()
    mdd   = ((cum - cum.cummax()) / cum.cummax()).min()
    return {
        'Portfolio': name,
        'Ann Ret Arith (%)': round(mu_a * 100, 2),
        'Ann Ret Geom (%)':  round(mu_g * 100, 2),
        'Ann Vol (%)':       round(vol * 100, 2),
        'Sharpe':            round(sr, 3),
        'Min Mo (%)':        round(rp.min() * 100, 2),
        'Max Mo (%)':        round(rp.max() * 100, 2),
        'Max DD (%)':        round(mdd * 100, 2),
    }

stats_df = pd.DataFrame([
    perf_stats(rp_vw, rf_mon, 'Value-Weighted'),
    perf_stats(rp_mv, rf_mon, 'Min-Variance'),
]).set_index('Portfolio')

stats_df
```

## Excel export

Fill in the official template + save extended results for reference.

```python
from openpyxl import load_workbook
from openpyxl.drawing.image import Image as XlImage

def tpl_stats(rp, rf_s):
    """Stats as decimals for the template cells."""
    rp = rp.dropna()
    rf = rf_s.reindex(rp.index).ffill().fillna(0)
    T = len(rp)
    mu_a = 12 * rp.mean()
    mu_g = (1 + rp).prod() ** (12/T) - 1
    vol  = rp.std() * np.sqrt(12)
    rf_a = 12 * rf.mean()
    sr   = (mu_a - rf_a) / vol
    return {
        'ann_avg_ret': mu_a,
        'ann_vol':     vol,
        'ann_cum_ret': mu_g,
        'sharpe':      sr,
        'min_mo':      rp.min(),
        'max_mo':      rp.max(),
    }

vw_s = tpl_stats(rp_vw, rf_mon)
mv_s = tpl_stats(rp_mv, rf_mon)
```

```python
xlsx_out = f'{OUT}SAAM_Part1_EUR_template.xlsx'

wb = load_workbook(TPL)
ws = wb['Sheet1']

# summary stats: col B = VW, col C = MV
keys = ['ann_avg_ret', 'ann_vol', 'ann_cum_ret', 'sharpe', 'min_mo', 'max_mo']
for i, k in enumerate(keys):
    ws.cell(row=3+i, column=2, value=round(vw_s[k], 8))
    ws.cell(row=3+i, column=3, value=round(mv_s[k], 8))

# cumulative return chart for the template
cum_path = f'{OUT}SAAM_Part1_EUR_cumulative.png'
fig2, ax2 = plt.subplots(figsize=(7, 4))
cm = (1 + rp_mv.dropna()).cumprod()
cv = (1 + rp_vw.dropna()).cumprod()
ax2.plot(cm.index, cm.values, color='steelblue', lw=1.8, label='Min-Var')
ax2.plot(cv.index, cv.values, color='darkorange', lw=1.8, ls='--', label='Val-Wgt')
ax2.set_title('Cumulative Return (base=1, Jan 2014)')
ax2.set_ylabel('Cumulative return')
ax2.legend(fontsize=9); ax2.grid(alpha=0.3)
ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax2.xaxis.set_major_locator(mdates.YearLocator(2))
fig2.tight_layout()
fig2.savefig(cum_path, dpi=150, bbox_inches='tight')
plt.close(fig2)

img = XlImage(cum_path)
img.width, img.height = 500, 280
ws.add_image(img, 'B9')

# monthly returns — match on (year, month) since DS dates != calendar month-end
vw_ym = {(d.year, d.month): v for d, v in rp_vw.items()}
mv_ym = {(d.year, d.month): v for d, v in rp_mv.items()}

for row in range(3, 3 + 144):
    dt_cell = ws.cell(row=row, column=5).value
    if dt_cell is None:
        continue
    dt = pd.Timestamp(dt_cell)
    ym = (dt.year, dt.month)
    vw_v = vw_ym.get(ym, np.nan)
    mv_v = mv_ym.get(ym, np.nan)
    ws.cell(row=row, column=6,
            value=round(float(vw_v), 8) if not np.isnan(vw_v) else None)
    ws.cell(row=row, column=7,
            value=round(float(mv_v), 8) if not np.isnan(mv_v) else None)

wb.save(xlsx_out)
print(f'Template saved: {xlsx_out}')
```
