## Loading CO2 data

```python
scope1 = pd.read_excel("Data_2026/DS_CO2_SCOPE_1_Y_2025.xlsx")
scope2 = pd.read_excel("Data_2026/DS_CO2_SCOPE_2_Y_2025.xlsx")

# drop rows with no ISIN or NAME (datastream errors)
scope1 = scope1.dropna(subset=["ISIN", "NAME"])
scope2 = scope2.dropna(subset=["ISIN", "NAME"])

# $$ER strings -> NaN
yr_cols = [c for c in scope1.columns if c not in ["NAME", "ISIN"]]
scope1[yr_cols] = scope1[yr_cols].apply(pd.to_numeric, errors="coerce")
yr_cols = [c for c in scope2.columns if c not in ["NAME", "ISIN"]]
scope2[yr_cols] = scope2[yr_cols].apply(pd.to_numeric, errors="coerce")

scope1_2 = pd.merge(scope1, static, on=["ISIN", "NAME"])
scope2_2 = pd.merge(scope2, static, on=["ISIN", "NAME"])

scope1_eu = scope1_2[scope1_2["Region"] == "EUR"].copy()
scope2_eu = scope2_2[scope2_2["Region"] == "EUR"].copy()

print(scope1_eu.shape, scope2_eu.shape)
```

## Cleaning price data

```python
# data_frames list: [yearly_ri_eu, monthly_ri_eu, yearly_mv_eu, monthly_mv_eu, rev_eu]
# all transposed so rows=dates, cols=firm names
# still may contain $$ER strings -> convert everything to numeric
for i in range(len(data_frames)):
    data_frames[i] = data_frames[i].apply(pd.to_numeric, errors="coerce")

yearly_ri  = data_frames[0].copy()
monthly_ri = data_frames[1].copy()
yearly_mv  = data_frames[2].copy()
monthly_mv = data_frames[3].copy()
rev        = data_frames[4].copy()

yearly_ri.index = yearly_ri.index.astype(int)
yearly_mv.index = yearly_mv.index.astype(int)
rev.index       = rev.index.astype(int)

monthly_ri.index = pd.to_datetime(monthly_ri.index)
monthly_mv.index = pd.to_datetime(monthly_mv.index)

print("yearly RI:", yearly_ri.shape, "  monthly RI:", monthly_ri.shape)
```

```python
# PDF says: treat RI < 0.5 as missing -> can lead to extreme/infinite returns otherwise
monthly_ri[monthly_ri < 0.5] = np.nan
yearly_ri[yearly_ri < 0.5]   = np.nan

print("monthly RI NaNs after low-price filter:", monthly_ri.isna().sum().sum())
```

## Computing monthly returns

```python
# simple return: (P_t / P_{t-1}) - 1
monthly_ret = monthly_ri.pct_change()

# first row NaN is expected
print(monthly_ret.shape)
monthly_ret.iloc[1:4, :4]
```

## Handling delistings

```python
# Datastream puts 'DEAD' in the name when a firm delists
# in that case, the price goes to 0 -> realized return = -100%
# we set the return to -1 in the first month after the last observed price

dead_firms = [col for col in monthly_ret.columns if "DEAD" in str(col).upper()]
print(f"{len(dead_firms)} delisted firms found")

for firm in dead_firms:
    last_px = monthly_ri[firm].last_valid_index()
    if last_px is None:
        continue
    after = monthly_ret.index[monthly_ret.index > last_px]
    if len(after) > 0:
        monthly_ret.loc[after[0], firm] = -1.0
```

## Stale price detection

```python
# if a firm has > 50% months with return = 0 over the estimation window, it's illiquid
# -> exclude from investment that year

stale_threshold = 0.50
alloc_years = list(range(2013, 2025))
stale_by_year = {}

for Y in alloc_years:
    w_end   = pd.Timestamp(f"{Y}-12-31")
    w_start = pd.Timestamp(f"{Y-10}-01-01")
    
    w = monthly_ret.loc[w_start:w_end]
    zero_frac = (w == 0).sum() / len(w)
    stale_by_year[Y] = set(zero_frac[zero_frac > stale_threshold].index.tolist())

print("avg stale firms per year:", round(np.mean([len(v) for v in stale_by_year.values()]), 1))
```

## CO2 data: forward fill gaps

```python
drop_cols = ["ISIN", "NAME", "Country", "Region"]

s1_t = scope1_eu.drop(columns=drop_cols, errors="ignore").set_index(scope1_eu["NAME"].values).T
s1_t.index = s1_t.index.astype(int)

s2_t = scope2_eu.drop(columns=drop_cols, errors="ignore").set_index(scope2_eu["NAME"].values).T
s2_t.index = s2_t.index.astype(int)

# ffill fills mid-sample and end-of-sample gaps, leaves beginning-of-sample NaN
s1_t = s1_t.ffill()
s2_t = s2_t.ffill()

# same for revenue
rev_ff = rev.ffill()

print("scope1 (ffilled):", s1_t.shape)
print("scope2 (ffilled):", s2_t.shape)
```

## Building investable universe per year

```python
min_monthly_obs = 36  # at least 3 years of data in 10yr window
investable = {}

for Y in alloc_years:
    w_end   = pd.Timestamp(f"{Y}-12-31")
    w_start = pd.Timestamp(f"{Y-10}-01-01")
    
    # 1) price available at end of year Y
    if Y in yearly_ri.index:
        has_price = set(yearly_ri.loc[Y].dropna().index)
    else:
        has_price = set()
    
    # 2) enough return observations
    w = monthly_ret.loc[w_start:w_end]
    obs = w.count()
    enough = set(obs[obs >= min_monthly_obs].index)
    
    # 3) not stale
    not_stale = enough - stale_by_year[Y]
    
    # 4) carbon data available (scope1 or scope2)
    firms_with_co2 = set()
    for firm in not_stale:
        s1_ok = (firm in s1_t.columns) and (Y in s1_t.index) and not pd.isna(s1_t.at[Y, firm])
        s2_ok = (firm in s2_t.columns) and (Y in s2_t.index) and not pd.isna(s2_t.at[Y, firm])
        if s1_ok or s2_ok:
            firms_with_co2.add(firm)
    
    investable[Y] = sorted(firms_with_co2 & has_price)

for Y in alloc_years:
    print(f"{Y}: {len(investable[Y])} investable firms")
```

## Summary

```python
print("Data after cleaning (EUR region):")
print(f"  yearly RI:   {yearly_ri.shape[1]} firms, {yearly_ri.shape[0]} years")
print(f"  monthly RI:  {monthly_ri.shape[1]} firms, {monthly_ri.shape[0]} months")
print(f"  monthly ret: {monthly_ret.shape[1]} firms")
print(f"  scope1:      {s1_t.shape[1]} firms")
print(f"  scope2:      {s2_t.shape[1]} firms")
print()
print(f"Investable universe size per year:")
for Y, lst in investable.items():
    print(f"  {Y}: {len(lst)} firms")
```
