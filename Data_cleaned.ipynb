```python
# Transpose so time is on rows, firms on columns.
# Drop the id/meta columns before transposing, then use the NAME row as column headers.
cols_to_drop = ["ISIN", "Country", "Region"]

frames_eur = [ri_annual_eur, ri_monthly_eur, mv_annual_eur, mv_monthly_eur, rev_eur]
frames_eur = [df.drop(columns=cols_to_drop, errors='ignore') for df in frames_eur]
frames_eur = [df.T for df in frames_eur]
frames_eur = [df.iloc[1:].set_axis(df.iloc[0], axis=1) for df in frames_eur]
```

## Handling CO2 data

Same pipeline as above: strip the bogus header row, force numeric types, merge with static, filter to EUR.

```python
# Drop the NaN-ISIN row (Datastream artifact)
co2_s1 = co2_s1.dropna(subset=["ISIN", "NAME"])
co2_s2 = co2_s2.dropna(subset=["ISIN", "NAME"])

yr_cols = [c for c in co2_s1.columns if c not in ["NAME", "ISIN"]]
co2_s1[yr_cols] = co2_s1[yr_cols].apply(pd.to_numeric, errors="coerce")

yr_cols = [c for c in co2_s2.columns if c not in ["NAME", "ISIN"]]
co2_s2[yr_cols] = co2_s2[yr_cols].apply(pd.to_numeric, errors="coerce")

s1_full = pd.merge(co2_s1, static_info, on=["ISIN", "NAME"])
s2_full = pd.merge(co2_s2, static_info, on=["ISIN", "NAME"])

s1_eur = s1_full[s1_full["Region"] == "EUR"].copy()
s2_eur = s2_full[s2_full["Region"] == "EUR"].copy()

s1_eur.shape, s2_eur.shape
```

## Cleaning price data

Force numeric ($$ER strings -> NaN), set proper index types for each frequency.

```python
for i in range(len(frames_eur)):
    frames_eur[i] = frames_eur[i].apply(pd.to_numeric, errors="coerce")

ri_y  = frames_eur[0].copy()   # annual RI
ri_m  = frames_eur[1].copy()   # monthly RI (used for return calculation)
mv_y  = frames_eur[2].copy()   # annual market cap
mv_m  = frames_eur[3].copy()   # monthly market cap
rev   = frames_eur[4].copy()   # revenues (thousands USD)

ri_y.index = ri_y.index.astype(int)
mv_y.index = mv_y.index.astype(int)
rev.index  = rev.index.astype(int)

ri_m.index = pd.to_datetime(ri_m.index)
mv_m.index = pd.to_datetime(mv_m.index)

ri_y.shape, ri_m.shape
```

#### Handling RI < 0.5

Prices below 0.5 are set to NaN as per the project instructions. These often correspond to rounding artifacts in Datastream where a stock that nearly went to zero has its RI recorded as 0 or some tiny float, which would generate infinite/extreme returns.

```python
ri_m[ri_m < 0.5] = np.nan
ri_y[ri_y < 0.5] = np.nan

ri_m.isna().sum().sum()   # total NaN count after cleaning
```

## Computing monthly returns

Simple returns: R_t = (P_t / P_{t-1}) - 1. Using pct_change() which does exactly that.

```python
ret_m = ri_m.pct_change()
ret_m.iloc[1:4, :4]   # quick sanity check on a few firms
ret_m.shape
```

## Handling delistings

Firms flagged as 'DEAD' by Datastream were delisted. At the first month after their last valid price observation, we assign a -100% return to capture the full capital loss for anyone who held through delisting.

```python
dead_firms = [col for col in ret_m.columns if "DEAD" in str(col).upper()]
print(f"{len(dead_firms)} delisted firms found")
# 58 delistings is quite a few -- this matters for portfolio performance in some years

for firm in dead_firms:
    last_valid = ri_m[firm].last_valid_index()
    if last_valid is None:
        continue
    dates_after = ret_m.index[ret_m.index > last_valid]
    if len(dates_after) > 0:
        ret_m.loc[dates_after[0], firm] = -1.0
```

## Stale price detection

For each allocation year, compute the fraction of zero-return months over the preceding 10-year estimation window. Firms above 50% get flagged as stale and excluded -- they would get massively over-weighted by the min-variance optimizer since their artificially low estimated variance makes them look like "free" risk reduction.

```python
STALE_THRESH = 0.50
alloc_years = list(range(2013, 2025))
stale_by_year = {}

for Y in alloc_years:
    end_of_year     = pd.Timestamp(f"{Y}-12-31")
    start_of_window = pd.Timestamp(f"{Y-10}-01-01")

    w = ret_m.loc[start_of_window:end_of_year]
    zero_frac = (w == 0).sum() / len(w)
    stale_by_year[Y] = set(zero_frac[zero_frac > STALE_THRESH].index.tolist())

# average number of stale firms per year
round(np.mean([len(v) for v in stale_by_year.values()]), 1)
```
